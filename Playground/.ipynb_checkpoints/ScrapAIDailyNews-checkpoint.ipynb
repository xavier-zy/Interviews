```python
import pandas as pd
```

```python
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait # available since 2.4.0
from selenium.webdriver.support import expected_conditions as EC # available since 2.26.0
import time
```

```python
from selenium.webdriver import ActionChains
```

```python
data = pd.DataFrame(columns = ["month", "day", "title", "content"])
```

```python
driver = webdriver.Chrome()
driver.get("https://www.jiqizhixin.com/dailies")

print(driver.title)

date = driver.find_elements_by_class_name("js-daily-every")
month = date[-1].find_elements_by_class_name("daily-every__month")
day = date[-1].find_elements_by_class_name("daily-every__day")
print(month[0].text, day[0].text)

month_text = month[0].text

flag = True
while flag:
    loadButton = driver.find_elements_by_class_name("u-loadmore__btn")
    ActionChains(driver).move_to_element(loadButton[0]).click().perform()
    time.sleep(1)
    date = driver.find_elements_by_class_name("js-daily-every")
    month = date[-1].find_elements_by_class_name("daily-every__month")
    day = date[-1].find_elements_by_class_name("daily-every__day")
    print("\n", month[0].text, day[0].text)

    month_text = month[0].text
    day_text = day[0].text
    if month_text == "2月" and day_text in ["20","21","22","23","24","25","26","27"]:
        news = date[-1].find_elements_by_class_name("daily-every__list")
        news_list = news[-1].find_elements_by_class_name("daily-every__item")
        for news in news_list:
            title = news.find_elements_by_class_name("daily-every__title")[0].text
            content = news.find_elements_by_class_name("daily__content")[0].text
            print(title)
            new_row = {
                "month": month_text,
                "day": day_text,
                "title": title,
                "content": content
            }
            data = data.append(new_row, ignore_index=True)
        
    if month_text == "2月" and day_text == "19":
        flag = False

driver.quit()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.to_csv("data.csv", index=False)
```
